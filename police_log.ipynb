```python
import pandas as pd
from datetime import datetime

#Step 1 Load data 
file_path = r"D:\Police_Log_checkpost\traffic_stops - traffic_stops_with_vehicle_number.csv"
df = pd.read_csv(file_path)
```

```python
df.dropna(axis=1, how="all", inplace=True)
```

```python
#Creating Database as secure_check
cursor.execute("create database if not exists secure_check;")

print("Database created sucessfully!")
```

```python
#Table creation 
cursor.execute("Use secure_check")
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS policelog (
      stop_date DATE,
      stop_time TIME,
      country_name VARCHAR(512),
      driver_gender VARCHAR(512),
      driver_age_raw  BIGINT(20),
      driver_age BIGINT(20),
      driver_race VARCHAR(512),
      violation_raw VARCHAR(512),
      violation VARCHAR(512),
      search_conducted VARCHAR(512),
      search_type VARCHAR(512),
      stop_outcome VARCHAR(512),
      is_arrested VARCHAR(512),
      stop_duration VARCHAR(512),
      drugs_related_stop VARCHAR(512),
      vehicle_number VARCHAR(512),
      timestamp DATETIME )
    
""")
conn.commit()

print("Table created sucessfully!")
```

```python
#Inserting values using placeholders
#data_list = df.values.tolist()

insert_query = """
    INSERT INTO policelog (stop_date,stop_time,country_name,driver_gender,driver_age_raw,driver_age,driver_race,violation_raw,violation,
    search_conducted,search_type,stop_outcome,is_arrested,stop_duration,drugs_related_stop,vehicle_number,timestamp)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
"""

# Loop through and insert each row
# ---------- Insert Row by Row ----------
for _, row in df.iterrows():
    data = (
        row.get('stop_date'),
        row.get('stop_time'),
        row.get('country_name'),
        row.get('driver_gender'),
        int(row['driver_age_raw']) if row.get('driver_age_raw') is not None else None,
        int(row['driver_age']) if row.get('driver_age') is not None else None,
        row.get('driver_race'),
        row.get('violation_raw'),
        row.get('violation'),
        row.get('search_conducted'),
        row.get('search_type'),
        row.get('stop_outcome'),
        row.get('is_arrested'),
        row.get('stop_duration'),
        row.get('drugs_related_stop'),
        row.get('vehicle_number'),
        row.get('timestamp')
    )

    cursor.execute(insert_query, data)

conn.commit()
print("✅ Data inserted successfully!")

# ---------- Close ----------
cursor.close()
conn.close()

```

```python
#Q8.Are stops during the night more likely to lead to arrests?
cursor.execute("""
        SELECT CASE
           WHEN HOUR(stop_time) BETWEEN 20 AND 23 THEN 'Night'
           WHEN HOUR(stop_time) BETWEEN 0 AND 5 THEN 'Night'
           ELSE 'Day'
       END AS time_of_day,
       COUNT(CASE WHEN stop_outcome = 'Arrest' THEN 1 END) * 100.0 / COUNT(*) AS arrest_rate
FROM policelog
GROUP BY time_of_day""")

out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
# q1. Yearly Breakdown of Stops and Arrests by Country (Using Subquery and Window Functions)
cursor.execute("""
        SELECT 
    country_name,
    year,
    total_stops,
    arrests,
    SUM(arrests) OVER (PARTITION BY country_name ORDER BY year) AS running_arrests
FROM (
    SELECT 
        country_name,
        YEAR(stop_time) AS year,
        COUNT(*) AS total_stops,
        COUNT(CASE WHEN is_arrested = 1 THEN 1 END) AS arrests
    FROM policelog
    GROUP BY country_name, year
) AS t
ORDER BY country_name, year;""")

out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#q2. Driver Violation Trends Based on Age and Race (Join with Subquery)
cursor.execute("""
        SELECT 
    p.driver_age,
    p.driver_race,
    p.violation,
    COUNT(*) AS stop_count
FROM policelog p
JOIN (
    SELECT driver_age, driver_race
    FROM policelog
    GROUP BY driver_age, driver_race
) AS t
ON p.driver_age = t.driver_age
AND p.driver_race = t.driver_race
GROUP BY p.driver_age, p.driver_race, p.violation
ORDER BY stop_count DESC;""")

out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#q3. Time Period Analysis of Stops (Joining with Date Functions) , Number of Stops by Year,Month, Hour of the Day
cursor.execute("""
       SELECT 
    YEAR(STR_TO_DATE(p.stop_date, '%Y-%m-%d')) AS year,
    MONTH(STR_TO_DATE(p.stop_date, '%Y-%m-%d')) AS month,
    CAST(SUBSTRING_INDEX(p.stop_time, ':', 1) AS UNSIGNED) AS hour,
    COUNT(*) AS stop_count
FROM policelog p
JOIN (SELECT 1 AS dummy) d ON 1 = d.dummy 
WHERE p.stop_date IS NOT NULL 
  AND p.stop_time IS NOT NULL
GROUP BY year, month, hour
ORDER BY year, month, hour;""")

out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#q4. Violations with High Search and Arrest Rates (Window Function)
cursor.execute("""
       SELECT 
    violation,
    search_rate,
    arrest_rate,
    RANK() OVER (ORDER BY (search_rate + arrest_rate) DESC) AS rank_order
FROM (
    SELECT 
        violation,
        COUNT(CASE WHEN search_conducted = 1 THEN 1 END) * 100.0 / COUNT(*) AS search_rate,
        COUNT(CASE WHEN is_arrested = 1 THEN 1 END) * 100.0 / COUNT(*) AS arrest_rate
    FROM policelog
    GROUP BY violation
) AS t
ORDER BY rank_order;""")

out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#q5. Driver Demographics by Country (Age, Gender, and Race)
cursor.execute("""
        SELECT country_name, driver_age, driver_gender, driver_race, COUNT(*) AS stop_count
        FROM policelog
        GROUP BY country_name, driver_age, driver_gender, driver_race
        ORDER BY country_name, stop_count DESC;
    """)

out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#q6. Top 5 Violations with Highest Arrest Rates
cursor.execute("""
SELECT violation, 
       COUNT(*) AS total, 
       SUM(CASE WHEN is_arrested = 1 THEN 1 ELSE 0 END) AS arrests,
       SUM(CASE WHEN is_arrested = 1 THEN 1 ELSE 0 END) * 1.0 / COUNT(*) AS arrest_rate
FROM policelog
GROUP BY violation
ORDER BY arrest_rate DESC
LIMIT 5""")

out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```
